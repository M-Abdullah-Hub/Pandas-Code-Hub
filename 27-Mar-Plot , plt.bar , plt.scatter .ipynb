```python
import numpy as np
import matplotlib.pyplot as plt

# Define the arrays with the same size
array3 = np.array([15, 19, 13, 14, 34, 70, 45, 65, 84, 42])
array4 = np.array([31, 90, 32, 78, 31, 71, 54, 20, 48, 29])

# Plot the scatter plot
plt.scatter(array1, array2)

# Show the plot
plt.show()
```

```python
plt.scatter(array1,array2,)
plt.scatter(array3, array4,s=100,marker='*',alpha=0.5,cmap='CMRmap')
plt.show()
```

```python
color=[1,9,2,8,1,7,5,2,8,3]
plt.scatter(array3,array4,s=100,c=color,alpha=0.5,cmap='coolwarm')
plt.colorbar()
plt.show()
```
